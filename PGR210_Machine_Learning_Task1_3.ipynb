```python
# Ref: “Plot Different SVM Classifiers in the Iris Dataset.” n.d. Scikit-Learn 0.18.2 Documentation. https://scikit-learn.org/0.18/auto_examples/svm/plot_iris.html.
# Ref: “Legend Guide.” n.d. Matplotlib 3.6.2 Documentation. https://matplotlib.org/stable/tutorials/intermediate/legend_guide.html.
# Ref: Singla, Aarohi. 2022. “Support Vector Machine - (SVM) -Classification Algorithm-Practical Example.” GitHub. December 26, 2022. https://github.com/AarohiSingla/Support-Vector-Machine/blob/main/svm.ipynb.

# Import necessary libraries to solve the question
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn import metrics
from sklearn import svm

# Define the function to load the dataset
def read_data():
  data = pd.read_csv("gender_classification_v7.csv")

  # The data type needs to be changed from 'Object' to 'Int' for further execution because of that we change the data type
  data['gender'].mask(data['gender'] == 'Male', 0, inplace = True)
  data['gender'].mask(data['gender'] == 'Female', 1, inplace = True)

  data['gender'] = data['gender'].astype('int')
 
  # Selecting numerical-continuous columns
  x = data.iloc[:, [1,2]]
  # Select columns for classification to use the SVM model
  y = data.iloc[:, 7]
  return x, y

# Define the function to formatting the SVM model
def model_format(x, y):
  x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
  
  # Feature Scaling
  sc_x = StandardScaler()
  x_train = sc_x.fit_transform(x_train)
  x_test = sc_x.transform(x_test)

  # Using Linear and RBF kernels to fitting the SVM model
  svc = svm.SVC(kernel = 'linear', C = 0.1).fit(x_train, y_train)
  rbf_svc = svm.SVC(kernel = 'rbf', gamma = 0.01, C = 0.1).fit(x_train, y_train)

  # Predicting the test set results
  y_pred_svc = svc.predict(x_test)
  y_pred_rbf = rbf_svc.predict(x_test)

  print('Accuracy Score: Linear Kernel, Parameter C is 0.1')
  print(metrics.accuracy_score(y_test, y_pred_svc))
  print('')
  print('Accuracy Score: Linear rbf, Parameter C is 0.1')
  print(metrics.accuracy_score(y_test, y_pred_rbf))
  print('')

  return x_test, y_test, svc, rbf_svc

# Define the main function
def main():
  x,y = read_data()
  x_test, y_test, svc, rbf_svc = model_format(x, y)
  plot_graph(x_test, y_test, svc, rbf_svc)

# Define the function to plotting the graph
def plot_graph(x_test,y_test, svc,rbf_svc):
  # Step size in the mesh
  h = .02 
  x_min, x_max = x_test[:, 0].min() - 1, x_test[:, 0].max() + 1
  y_min, y_max = x_test[:, 1].min() - 1, x_test[:, 1].max() + 1
  xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))

  plt.scatter(x_test[:,0],x_test[:,1], c = y_test)

  titles = ['SVC with linear kernel C = 0.1',
          'SVC with rbf kernel C = 0.1',]

  for i, clf in enumerate((svc, rbf_svc)):
      # Plot the decision boundary. For that, we will assign a color to each point in the mesh [x_min, x_max]x[y_min, y_max].
      plt.subplot(2, 2, i + 1)
      plt.subplots_adjust(wspace=0.4, hspace=0.4)

      Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])

      # Put the result into a color plot
      Z = Z.reshape(xx.shape)
      plt.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.8)

      # Plot also the training points
      plt.scatter(x_test[:, 0], x_test[:, 1], c=y_test, cmap=plt.cm.coolwarm)
      plt.xlim(xx.min(), xx.max())
      plt.ylim(yy.min(), yy.max())
      blue_patch = mpatches.Patch(color='royalblue', label='Male')
      red_patch = mpatches.Patch(color='firebrick', label='Female')
      plt.legend(handles=[blue_patch, red_patch], fontsize = 'small')
      plt.xticks(())
      plt.yticks(())
      plt.title(titles[i], bbox={'facecolor':'0.8', 'pad':5}, fontsize = 15, pad = 30)

      plt.show()

if __name__ == '__main__':
    main()
```

### Getting Cross Validation Table

```python
# Guido, S., Müller,A.C. (2016). “Introduction to Machine Learning with Python”, Chapter 5, Page 273,274

# Define the function to load the dataset
def read_data():
  data = pd.read_csv("gender_classification_v7.csv")

  # The data type needs to be changed from 'Object' to 'Int' for further execution because of that we change the data type
  data['gender'].mask(data['gender'] == 'Male', 0, inplace = True)
  data['gender'].mask(data['gender'] == 'Female', 1, inplace = True)

  data['gender'] = data['gender'].astype('int')
 
  # Selecting every columns except for classification column
  x = data.iloc[:, [0,1,2,3,4,5,6]]
  # Select columns for classification to use the SVM model
  y = data.iloc[:, 7]
  return x, y

# Define the function to get a table of differences by kernel used
def get_table_kernel(x, y):
  x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
  
  # Feature Scaling
  sc_x = StandardScaler()
  x_train = sc_x.fit_transform(x_train)
  x_test = sc_x.transform(x_test)

  param_grid = [{'kernel': ['rbf'],
  'C': [0.1, 1, 100],
  'gamma': [0.01, 0.1, 1, 10, 100]},
  {'kernel': ['linear'],
  'C': [0.1, 1, 100]}]
  print("List of grids:\n{}".format(param_grid))

  # Grid search involving kernel and parameters 
  grid_search = GridSearchCV(SVC(), param_grid, cv=5)
  grid_search.fit(x_train, y_train)
  print("Best parameters: {}".format(grid_search.best_params_))
  print("Best cross-validation score: {:.2f}".format(grid_search.best_score_))

  results = pd.DataFrame(grid_search.cv_results_)
  return results

def main():
  x,y = read_data()
  results = get_table_kernel(x,y)
  display(results.T)

if __name__ == '__main__':
    main()
```
